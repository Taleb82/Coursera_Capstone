```python
import folium
# create map of Toronto using latitude and longitude values
map_toronto= folium.Map(location=[43.8, -79.2], zoom_start=10)

# add markers to map
for lat, lng, poscode in zip(NT_T['Latitude'], NT_T['Longitude'], NT_T['Postal Code']):
    label = '{}'.format(poscode)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
